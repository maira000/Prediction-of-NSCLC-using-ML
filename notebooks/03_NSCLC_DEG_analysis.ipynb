```r
# Step 1: Read CSV into a data frame
file_path <- '../data/raw/data_R.csv'
df <- read.csv(file_path, row.names = 1)
```

```r
# Preview the data
head(df)
```

```r
names(df)
```

```r
# Check for duplicates in the first column (if any)
duplicated(df$X)
length(df$X)
length(unique(df$X))
```

```r
# Step 2: Set row names (gene IDs)
row.names(df) <- df$X

# Remove the redundant column if it exists
if("X" %in% names(df)) {
  df <- df[, !(names(df) %in% "X")]
}

# Make sure all counts are numeric
df[] <- lapply(df, as.numeric)

# Preview cleaned data
head(df)
dim(df)
```

```r
library(edgeR)

# Step 3: Define the experimental groups
# Replace the numbers with your sample counts (T = tumor, N = normal)
expt <- rep(c("T", "N"), c(199, 19))  
expt <- factor(expt, levels = c("T", "N"))  # ensure order

# Step 3a: Replace negative counts with 0
rawcounts <- df
rawcounts[rawcounts < 0] <- 0

# Step 3b: Compute column sums (library sizes)
sum_size <- colSums(rawcounts)
sum_size

# Step 3c: Create DGEList
dge <- DGEList(counts = as.matrix(rawcounts), group = expt, lib.size = sum_size)

# Step 3d: Normalize counts using TMM
dge <- calcNormFactors(dge)

# Step 3e: Check normalized sample info
dge$samples
```

```r
library(edgeR)

# Step 4a: MDS plot
plotMDS(dge, main = "MDS plot of samples", labels = colnames(dge))

# Step 4b: Estimate common dispersion
dge <- estimateCommonDisp(dge)

# Step 4c: Check DGEList after dispersion
dge$samples
```

```r
# Step 5: Differential expression
nbt <- exactTest(dge, pair = c("N", "T"))

# View results
nbt$table

# FDR correction for multiple testing
nbt.corrected <- topTags(nbt, n = Inf)
nbt.corrected$table

# Count significant genes at FDR < 0.05
sum(nbt.corrected$table$FDR < 0.05)
```

```r
# Step 6: Save the top DEGs
nbttable <- nbt.corrected$table

# Select significant genes at FDR < 0.05
deg <- nbttable[nbttable$FDR < 0.05, ]
deg$ID <- rownames(deg)  # Add gene IDs as a column

# Separate up-regulated genes (logFC > 0)
deg_up <- deg[deg$logFC > 0, ]
deg_up <- deg_up[order(deg_up$logFC, decreasing = TRUE), ]  # Sort by logFC descending

# Separate down-regulated genes (logFC < 0)
deg_down <- deg[deg$logFC < 0, ]
deg_down <- deg_down[order(deg_down$logFC), ]  # Sort by logFC ascending

# Save to CSV
write.csv(deg_up, file = "../data/processed/UPRegulated_NSCLC.csv", row.names = FALSE)
write.csv(deg_down, file = "../data/processed/DownRegulated_NSCLC.csv", row.names = FALSE)

# Quick check
dim(deg_up)
dim(deg_down)
head(deg_up)
head(deg_down)
```

```r
# Add row names as a column in rawcounts
rawcounts$ID <- rownames(rawcounts)

# Merge raw counts with all significant DEGs
combined <- merge(rawcounts, deg, by = "ID", all.y = TRUE)  # Keep only DEGs

# Sort combined data by logFC descending (optional, for visualization)
combined <- combined[order(combined$logFC, decreasing = TRUE), ]

# Quick check
head(combined)
dim(combined)
```

```r
library(dplyr)

# Ensure numeric values for heatmap (exclude non-numeric columns)
numeric_combined <- combined %>%
  mutate(across(where(is.numeric), as.numeric),  # Convert numeric columns to numeric
         across(where(is.character), as.character))  # Keep character columns as character

# Extract gene IDs
gene_ids <- numeric_combined$ID

# Extract only the counts for heatmap (exclude logFC, PValue, FDR)
# Assuming your raw counts columns are the first N columns before logFC/FDR
count_columns <- colnames(rawcounts)  # names of count columns
numeric_values <- as.matrix(numeric_combined[, count_columns])  # matrix for heatmap

# Quick check
str(numeric_values)
dim(numeric_values)
```

```r
# ------------------------------
# Prepare DEGs numeric matrix
# ------------------------------

# Values for T1-T199 and N1-N19 (combine into one matrix)
# For brevity, I'm using a simplified example; replace '...' with all your actual numbers
values <- c(
  669027.629702579,1026118.76411398,609899.34899679,692168.505136036,646776.890184368,518100.817370064,
  674198.148431355,732084.889819032,667029.791987507,721153.328290691,666472.977660902,652832.958919618,
  # ... include all T1-T199
  640925.030486726,622284.581975439,642350.156461527,625465.288106605,613750.858369929,
  655345.388827442,646261.539359595,635862.318857078,726433.625731905,681462.347855775
  # ... include all N1-N19
)

# Create a matrix
# Rows: genes (T1-T199), Columns: samples (N1-N19)
# Assuming you want genes as rows and samples as columns
gene_ids <- paste0("T", 1:199)
sample_ids <- paste0("N", 1:19)

# Make numeric matrix
numeric_matrix <- matrix(values, nrow = length(gene_ids), ncol = length(sample_ids), byrow = FALSE)
rownames(numeric_matrix) <- gene_ids
colnames(numeric_matrix) <- sample_ids

# ------------------------------
# Plot heatmap
# ------------------------------

heatmap(numeric_matrix, 
        Rowv = NA, Colv = NA,          # no clustering
        main = "Heatmap of DEGs in NSCLC", 
        col = heat.colors(256),
        scale = "row")
```

```r
# ------------------------------
# Save heatmap as PNG
# ------------------------------

# Open PNG device
png(filename = "../results/DEGs_heatmap.png",        # output file name
    width = 1200, height = 800,           # size in pixels
    res = 150)                            # resolution

# Plot heatmap
heatmap(numeric_matrix, 
        Rowv = NA, Colv = NA,             # no clustering
        main = "Heatmap of DEGs in NSCLC", 
        col = heat.colors(256),
        scale = "row")

# Close the device
dev.off()
```

```r
library(pheatmap)
library(RColorBrewer)

# Step 1: Calculate row variance
gene_variance <- apply(numeric_matrix, 1, var)

# Step 2: Select top 50 most variable genes
top_genes <- names(sort(gene_variance, decreasing = TRUE))[1:50]
numeric_matrix_top <- numeric_matrix[top_genes, ]

# Step 3: Define colors
heat_colors <- colorRampPalette(brewer.pal(n = 9, name = "YlOrRd"))(100)

# Step 4: Save heatmap as PNG
png("../results/Top50_DEGs_heatmap.png", width = 1600, height = 1200, res = 200)
pheatmap(numeric_matrix_top,
         cluster_rows = TRUE,
         cluster_cols = TRUE,
         show_rownames = TRUE,
         show_colnames = TRUE,
         color = heat_colors,
         scale = "row",
         main = "Top 50 Most Variable DEGs",
         fontsize = 10)
dev.off()

# Step 5: Save as PDF
pdf("../results/Top50_DEGs_heatmap.pdf", width = 12, height = 10)
pheatmap(numeric_matrix_top,
         cluster_rows = TRUE,
         cluster_cols = TRUE,
         show_rownames = TRUE,
         show_colnames = TRUE,
         color = heat_colors,
         scale = "row",
         main = "Top 50 Most Variable DEGs",
         fontsize = 10)
dev.off()

cat("Top 50 DEGs heatmap saved as PNG and PDF!\n")
```
